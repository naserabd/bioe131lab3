```python
import matplotlib.pyplot as plt
import string

#read contents
f = open("romeoandjuliet.txt", "r")
contents = f.read()

#remove punctuation
contents = contents.translate(str.maketrans("", "", punctuation)).lower()

#split contents into words
words = contents.split()

#Correct for present/past tense/plural
#from nltk.stem.snowball import SnowballStemmer
#stemmer = SnowballStemmer("english")
#Correction = [stemmer.stem(word) for word in words]
#print(Correction)
#Problem: It does not always stem correctly. Ex: stems "houses" to "hous"...

#create dictionary with words as keys and multiplicity as values
dictionary = {}
for word in words:
    if word in dictionary:
        dictionary[word] += 1 #dictionary[word] = dictionary[word] + 1 
    else:
        dictionary[word] = 1

#sort dictionary based on descending multiplicity and create list [word, multiplicity, rank]
sorted_list = [[x, dictionary[x]] for x in sorted(dictionary, key=lambda x: dictionary[x], reverse = True)]
[sorted_list[x].append(x+1) for x in range(0,len(sorted_list))]
print(sorted_list)
```

```python
plt.figure(figsize=(15,6))
#Plotting all the words takes forever for Jupyter, so only the most frequent 30 words are plotted.
plt.bar([sorted_list[x][0] for x in range(0,30)],[sorted_list[x][1] for x in range(0,30)],color = ['blue'])
plt.xlabel('Word')
plt.ylabel('Word Count')
```

```python
Top_Ten_Words = [sorted_list[x][0] for x in range(0,10)]
Least_Ten_Words = [sorted_list[-x-1][0] for x in range(0,10)]
print ("10 Most Frequent Words:")
print (Top_Ten_Words)
print ("10 Least Frequent Words:")
print (Least_Ten_Words)
```
